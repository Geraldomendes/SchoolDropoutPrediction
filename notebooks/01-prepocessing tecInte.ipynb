```python
data1 = pd.read_csv('../data/processed/preprocessingtecnicos.csv')
data1.head()
```

```python
data1.drop(columns=['Eixo Tecnológico', 'Tipo de Curso', 'Tipo Oferta'],axis=1, inplace=True)
```

```python
data1.drop(columns=['Situação de Matrícula'],axis=1, inplace=True)
```

```python
data1
```

```python
data1 ['Cor/Raça'].unique()
```

```python
data1 ['Cor/Raça'].str.lower()
```

```python
data1 ['Cor/Raça'] = data1 ['Cor/Raça'].str.lower()
```

```python
data1 ['Cor/Raça'].unique()
```

```python
data1 ['Instituição'].unique()
```

```python
data1 ['Instituição'].replace('IF SERTAO-PE', 'IF SERTÃO-PE')
```

```python
data1
```

```python
data1 ['Sexo'].unique()
```

```python
data1 ['Sexo'] = data1 ['Sexo'].replace('F', 'Feminino')
```

```python
data1 ['Sexo'] = data1 ['Sexo'].replace('M', 'Masculino')
```

```python
data1 ['Renda Familiar'].unique()
```

```python
data1 ['Renda Familiar'] = data1 ['Renda Familiar'].replace('1<RFP<=1,5', '1,0<RFP<=1,5')
```

```python
data1 ['Renda Familiar'] = data1 ['Renda Familiar'].replace('0,5<RFP<=1', '0,5<RFP<=1,0')
```

```python
data1 ['Renda Familiar'] = data1 ['Renda Familiar'].replace('Não Declarada', 'Não declarada')
```

```python
data1 ['Unidade de Ensino'].unique()
```

```python
data1['Categoria da Situação'].value_counts()
```

```python
data1['Fator Esforço Curso'] = data1['Fator Esforço Curso'].str.replace(',','.').astype(float)
```

```python
data1
```

```python
data1['Data de Início do Ciclo'] = pd.to_datetime(data1['Data de Início do Ciclo'], format="%d/%m/%Y")
data1['Data de Fim Previsto do Ciclo'] = pd.to_datetime(data1['Data de Fim Previsto do Ciclo'], format="%d/%m/%Y")
data1['Mês de Ocorrência da Situação'] = pd.to_datetime(data1['Mês de Ocorrência da Situação'], format="%d/%m/%Y")
```

```python
data1
```

```python
data1.to_csv(index=False, path_or_buf='preprocessingTEC.csv')
```
