```python
data1.drop(columns=['Eixo Tecnológico', 'Tipo de Curso', 'Situação de Matrícula'],axis=1, inplace=True)
```

```python
data1 ['Cor/Raça'].unique()
```

```python
data1 ['Cor/Raça'].str.lower()
```

```python
data1 ['Cor/Raça'] = data1['Cor/Raça'].str.lower()
```

```python
data1 ['Cor/Raça'].unique()
```

```python
data1 ['Instituição'].unique()
```

```python
data1 ['Instituição'].replace('UFSM - POLITECNICO', 'UFSM - POLITÉCNICO')
```

```python
data1 ['Instituição'] = data1 ['Instituição'].replace('UFSM - POLITECNICO', 'UFSM - POLITÉCNICO')
```

```python
data1
```

```python
data1 ['Sexo'].unique()
```

```python
data1 ['Sexo'] = data1 ['Sexo'].replace('F', 'Feminino')
```

```python
data1 ['Sexo'] = data1 ['Sexo'].replace('M', 'Masculino')
```

```python
data1 ['Renda Familiar'].unique()
```

```python
data1 ['Renda Familiar'] = data1 ['Renda Familiar'].replace('1<RFP<=1,5', '1,0<RFP<=1,5')
```

```python
data1 ['Renda Familiar'] = data1 ['Renda Familiar'].replace('0,5<RFP<=1', '0,5<RFP<=1,0')
```

```python
data1 ['Renda Familiar'] = data1 ['Renda Familiar'].replace('Não Declarada', 'Não declarada')
```

```python
data1 ['Unidade de Ensino'].unique()
```

```python
data1 ['Unidade de Ensino'] = data1 ['Unidade de Ensino'].replace('Colégio Politécnico da Universidade Federal Santa Maria', 'Colégio Politécnico da UFSM')
```

```python
data1.to_csv(index=False, path_or_buf='../data/processed/preprocessing2.csv')
```
